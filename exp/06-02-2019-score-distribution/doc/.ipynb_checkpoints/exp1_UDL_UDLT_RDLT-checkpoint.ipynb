```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gzip
import io
import math
```

```python
# Reading the results for a given value of k and n in model m with s samples

def read_results(k,n,m,s):
    RESULTS_SPECIES_aux = {}
    RESULTS_SCORES_aux  = {}
    with gzip.open('../results/samples_'+str(k)+'_'+str(n)+'_'+m+'.gz', 'r') as f:
        reader = csv.reader(io.TextIOWrapper(f, newline=""),delimiter='\t')        
        for row in reader:
            if row[0][0] == '#': #Species tree
                treeIndex = int(row[0].replace('# S_',''))
                RESULTS_SPECIES_aux[treeIndex] = {}
                RESULTS_SCORES_aux[treeIndex]  = {'D':[], 'L':[], 'T':[]}
                sampleIndex = 0
            else:
                if sampleIndex == 0:
                    for i in range(0,k):
                        RESULTS_SPECIES_aux[treeIndex][i] = []
                for i in range(0,k):
                    row1 = row[0].rstrip().split(' ')
                    count   = int(row1[i].split(':')[1])
                    RESULTS_SPECIES_aux[treeIndex][i].append(count)
                RESULTS_SCORES_aux[treeIndex]['D'].append(int(row1[k].split(':')[1]))
                RESULTS_SCORES_aux[treeIndex]['L'].append(int(row1[k+1].split(':')[1]))
                RESULTS_SCORES_aux[treeIndex]['T'].append(int(row1[k+2].split(':')[1]))
                sampleIndex += 1
                
    RESULTS_SPECIES = pd.DataFrame.from_dict({(t): RESULTS_SPECIES_aux[t]
                                            for t in RESULTS_SPECIES_aux.keys()},
                                            orient='index')
    RESULTS_SCORES = pd.DataFrame.from_dict({(t,e): RESULTS_SCORES_aux[t]
                                            for t in RESULTS_SCORES_aux.keys()
                                            for e in RESULTS_SCORES_aux[t].keys()},
                                            orient='index')
    return((RESULTS_SPECIES,RESULTS_SCORES))

def plot_species(df,i,w,h):
    df_i = df.iloc[i]
    Tot  = len(df_i)
    Cols = 3
    Rows = Tot // Cols 
    Rows += Tot % Cols
    Position = range(1,Tot + 1)
    fig = plt.figure(figsize=(w,h))
    plt.tight_layout()
    k=0
    for column in df:
        ax = fig.add_subplot(Rows,Cols,Position[k])
        plt.hist(np.array(df_i[column]),bins=max(df_i[column]), rwidth=0.5)
        plt.title('Tree '+str(i)+' species '+str(column))
        k+=1
    plt.show()
        
def plot_scores(df,i,w,h,T=False):
    df_i = df.iloc[i]
    fig = plt.figure(figsize=(w,h))
    plt.tight_layout()
    axD = fig.add_subplot(1,3,1)
    plt.hist(np.array(df_i['D']),bins=max(df_i['D']), rwidth=0.5)
    plt.title('Tree '+str(i)+' score D')
    axL = fig.add_subplot(1,3,2)
    plt.hist(np.array(df_i['L']),bins=max(df_i['L']), rwidth=0.5)
    plt.title('Tree '+str(i)+' score L')
    if T:
        axT = fig.add_subplot(1,3,3)
        plt.hist(np.array(df_i['T']),bins=max(df_i['T']), rwidth=0.5)
        plt.title('Tree '+str(i)+' score T')
    plt.show()
```

```python
(UDL_8_10_SPECIES,UDL_8_10_SCORES)   = read_results(8,10,'U_DL',10000)
(UDLT_8_10_SPECIES,UDLT_8_10_SCORES) = read_results(8,10,'U_DLT',10000)
(RDLT_8_10_SPECIES,RDLT_8_10_SCORES) = read_results(8,10,'R_DLT',10000)
```

```python
plot_species(UDL_8_10_SPECIES,0,20,15)
plot_scores(UDL_8_10_SCORES,0,20,3,False)
```

```python
plot_species(UDL_8_10_SPECIES,25,20,15)
plot_scores(UDL_8_10_SCORES,25,20,3,False)
```

```python
plot_species(UDLT_8_10_SPECIES,25,20,15)
plot_scores(UDLT_8_10_SCORES,25,20,3,True)
```

```python
plot_species(RDLT_8_10_SPECIES,25,20,15)
plot_scores(RDLT_8_10_SCORES,25,20,3,True)
```
